```python
import pandas as pd
df = pd.read_csv('./data/黑龙江/csv/哈尔滨公交站点.csv')
# 根据站点名称分组，并对name字段进行合并，使用', '作为分隔符
# 同时创建一个新字段'线路频次'，初始化为1
df['线路频次'] = 1
df = df.groupby('站点名称').agg({
    'lng': 'first',
    'lat': 'first',
    '正反线路': 'first',
    'name': lambda x: ', '.join(x),
    '线路频次': 'sum'
}).reset_index()
df
```
